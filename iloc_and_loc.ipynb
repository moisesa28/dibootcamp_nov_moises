```python
import pandas as pd
import seaborn as sns
```

```python
flights = sns.load_dataset("flights")
```

#loc and iloc - Precise selections

```python
flights['passengers']
```

```python
#how to select a specific row and column
#or change a specific value of data
```

```python
#iloc is position based by number
#example dataframe.iloc[0,2] will return the first row and third column
```

```python
#loc is label based
#example df.loc[0,'name'] this will return the first row of the name column
```

#.iloc

```python
flights.iloc[0]
```

```python
flights.iloc[0,2]
```

```python
flights.iloc[0:5]
```

```python
flights.iloc[0:5,0:2]
```

```python
#to get specific rows (more than 1)
```

```python
flights.iloc[0:11:5] #this is a loop (start is 0, end is 11 and the step is 5)
```

```python
#another aproach for specific rows: rows 2, 11 and 101
wanted_rows = [2,11,101]
flights.iloc[wanted_rows]
```

```python
wanted_columns = [1,2]
flights.iloc[wanted_rows,wanted_columns]
```

#.loc

```python
#specify the row and the column
#uses labels and column names
```

```python
flights.loc[0]
```

```python
flights.loc[0,'passengers']
```

```python
wanted_columns = ['passengers','year']
flights.loc[0:5,wanted_columns]
```

#combining boolean masks with loc:
it will only work with loc, not iloc

```python
high_traffic_mask = flights['passengers']>500
```

```python
flights.loc[high_traffic_mask,'year']
```

```python
flights.loc[high_traffic_mask,'year'].unique()
```

```python
mediumhigh_traffic_mask = flights['passengers']>400
after_1955_flights_mask = flights['year']>1955
flights.loc[(mediumhigh_traffic_mask) & (after_1955_flights_mask),wanted_columns]
```

```python
#get the summer months: Jun, Jul, Aug
#show only the month and passengers using loc
summer = ['Jun','Jul','Aug']
summer_mask = flights['month'].isin(summer)
wanted_columns = ['month','passengers']
flights.loc[summer_mask,wanted_columns]
```

```python
flights.loc[summer_mask,wanted_columns].head(10)
```

#Where iloc is powerful

setting values with iloc and loc

```python
flights_copy = flights.copy() #it will copy the dataset
```

```python
flights_copy
```

```python
flights_copy.loc[0,'passengers'] = 999
flights_copy.head()
```

#Adding values

```python
#to create a new column: 'traffic'=low, medium, high
```

```python
low_traffic_mask = flights_copy['passengers'].between(0,200)
medium_traffic_mask = flights_copy['passengers'].between(201,400)
high_traffic_mask = flights_copy['passengers']>400
```

```python
flights_copy.loc[low_traffic_mask,'traffic']='Low'
flights_copy.loc[medium_traffic_mask,'traffic']='Medium'
flights_copy.loc[high_traffic_mask,'traffic']='High'
flights_copy
```

```python
flights_copy.iloc[50:60]
```
